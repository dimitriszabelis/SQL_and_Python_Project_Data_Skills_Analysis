```python
# Read the csv file
import pandas as pd
file_path = '../data/b_sectors.csv'
df = pd.read_csv(file_path)
df
```

```python
# Change to ascending order (the graph is populated from bottom up)
import matplotlib.pyplot as plt
import numpy as np

sorted_df = df.sort_values(by='market_share', ascending=True)

# Horizontal Bar Chart Design

# Figure Size & Palette
plt.figure(figsize=(12, 6))
# Create a color gradient scheme for the different skills
gradient_colors = plt.cm.Blues(np.linspace(0.5, 1.1, 15))  # Blue gradient

# Create the horizontal bar chart
plt.barh(sorted_df['sector_name'], sorted_df['market_share'],color=gradient_colors)

# Market Share = The total number of jobs offered by companies of the same sector over 
# the total number of jobs offered by companies that offer 25 or more Senior Data Scientist jobs. 

# Add labels and title 
plt.xlabel('Market Share (%)')
plt.title('Market Share by Sector for Senior Data Scientist Jobs')

# Display percentage values on the bars
for index, value in enumerate(sorted_df['market_share']): # enumerate() returns the index and value for each entry in the 'market_share' column
    plt.text(value + 0.5, index, f'{value:.1f}%', va='center') #For each sector it places a text label on the plot
    # X-Axis Position = value + 0.5 (to offset the text from the bar)
    # Y-Axis Position = index (at the position of each sector on the y-axis)
    #Label = f'{value:.1f}%' (percentage formatted to 1 decimal place followed by a percent sign)
    #Alignment = va='center' (center the text vertically relative to its y-coordinate)

# Show the plot
plt.xlim(0, max(sorted_df['market_share']) + 2.5)  # Adjust x-axis limit for better visibility
plt.grid(axis='x', linestyle='--', alpha=0.6) #Grid along the x-axis, dashed lines, and transparency level to 0.7 (not fully opaque)
plt.tight_layout()  # Adjust layout to prevent clipping of labels (optimizes the layout of the elements within the figure)
plt.savefig("../assets/sectors.png", format='png', dpi=300, bbox_inches='tight') #tight layout around the figure (ensures that any extra whitespace around the figure is minimized when saving it)
plt.show()
```
