```python
# Read the csv file
import pandas as pd
file_path = 'data/ii2_top_paying_skills_having_without_duplicates_all_skills.csv'
# Read the first 15 lines of the CSV file, including the header
df_all_skills = pd.read_csv(file_path,nrows=15) #nrows = Number of data rows to read from the start of the file. The counting starts at 1.
print(df_all_skills)
```

```python
# Change to ascending order (the graph is populated from bottom up)
import matplotlib.pyplot as plt
import numpy as np

sorted_df = df_all_skills.sort_values(by='avg_salary_per_skill', ascending=True)

# Horizontal Bar Chart Design

# Figure Size & Palette
plt.figure(figsize=(12, 6))
# Create a color gradient scheme for the different skills
gradient_colors = plt.cm.Blues(np.linspace(0.5, 1.1, 15))  # Blue gradient

# Create the horizontal bar chart
plt.barh(sorted_df['skill_name'], sorted_df['avg_salary_per_skill'],color=gradient_colors)

# Add labels and title
plt.xlabel('Annual Salary ($)')
plt.title('Top-Paying Skills for the Top-Paying Senior Data Scientist Jobs')

# Customize the x-axis 
ticks = np.arange(0,225_000+50_000,50_000)
labels = [f'{x:,.0f}' for x in ticks] # Create tick labels
plt.xticks(ticks, labels)
plt.xlim([0, 225_000])
plt.grid(axis='x', linestyle='--', alpha=0.6) #Grid along the x-axis, dashed lines, and transparency level to 0.7 (not fully opaque)

# Display salary values close to the bars
for index, value in enumerate(sorted_df['avg_salary_per_skill']): # enumerate() returns the index and value for each entry in the 'avg_salary_per_skill' column
    plt.text(value + 2_000, index, f'${value:,.0f}', va='center') #For each skill it places a text label on the plot
    # X-Axis Position = value + 1_000 (to offset the text from the bar)
    # Y-Axis Position = index (at the position of each skill on the y-axis)
    #Label = f{value:,.0f} (no decimal places are shown and addition of the thousand separator)
    #Alignment = va='center' (center the text vertically relative to its y-coordinate)

# Create the Bar Chart
plt.tight_layout()  # Adjust layout to prevent clipping of labels (optimizes the layout of the elements within the figure)
plt.savefig("assets/top_paying_skills_bar_charts.pdf", format='pdf', dpi=300, bbox_inches='tight') #tight layout around the figure (ensures that any extra whitespace around the figure is minimized when saving it)
plt.show()
```
